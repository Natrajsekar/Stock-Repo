```python
def multiply(amount,status):
    if status=='Sale':
        return amount * 100000
    else:
        return None
    
sale_out_go['new']=sale_out_go.apply(lambda row: multiply(row['Sale Amount'],row['Final Status']),axis=1)
```

```python
sale_out_go
```

```python
sale_out_go['Valuation Price']=sale_out_go['Valuation Price'].replace('-',np.nan)
```

```python
sale_out_go['Valuation Price']=pd.to_numeric(sale_out_go['Valuation Price'],errors='coerce')
```

```python
sale_out_go['Valuation %']=(sale_out_go['new'] / sale_out_go['Valuation Price'])*100
```

```python
sale_out_go['Valuation %']=sale_out_go['Valuation %'].map(lambda x: '{:.2f}%'.format(x) if not np.isnan(x) else '')
```

```python
sale_out_go['Valuation %']=sale_out_go['Valuation %'].fillna('-')
sale_out_go['Valuation Price']=sale_out_go['Valuation Price'].fillna('-')
```
